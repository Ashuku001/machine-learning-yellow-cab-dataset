##### transformation pipelines

Crete transformation pipelines for the imputers used

```python
from sklearn.pipeline import make_pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
```

##### numerical transformation pipeline

```python
# mean imputerattributes = ['passenger_count', 'congestion_surcharge']
num_attributes = [ 'passenger_count', 'trip_distance','PULocationID', 'DOLocationID', 'fare_amount', 'extra', 'mta_tax',
       'tip_amount', 'tolls_amount', 'improvement_surcharge', 'total_amount', 'congestion_surcharge', 'tavg', 'tmin', 'tmax',
       'wspd', 'pres', 'pickup_weekday', 'pickup_hour', 'pickup_minute']
num_pipeline = make_pipeline(SimpleImputer(strategy="mean"),
                              StandardScaler())
```

##### categorical transformation pipeline

```python
cat_attributes = ['store_and_fwd_flag_N', 'store_and_fwd_flag_Y',]
cat_pipeline = make_pipeline(SimpleImputer(strategy="most_frequent"), 
                                       OneHotEncoder(handle_unknown="ignore"))
```

##### Combine the tranformers

```python
# name of tranformer, the tranformer itself and the list of columns the tranformer is applied to
preprocessing = ColumnTransformer([
        ("numerical", num_pipeline, num_attributes),
        ("categorical", cat_pipeline, cat_attributes),
    ])
```

#### applying tranformer to the whole dataset

[MUST] ensure the trips types are casted to numerical typs or categorical tyops where necessary

```python
trips_prepared = preprocessing.fit_transform(trips)
df_trips_prepared = pd.DataFrame(
                                trips_prepared, 
                                columns=preprocessing.get_feature_names_out(),
                                index=trips.index
                            )
df_trips_prepared.head()
```

This transformer will impute the missing values for numerical typs using mean and categorical will be onehotencoded.

##### save the prepared data for ML algorithms


```python
df_trips_prepared["trip_duration"] = trips_label
df_trips_prepared.to_parquet("data/trips-prepared-data")
```
