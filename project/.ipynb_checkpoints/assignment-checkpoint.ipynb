```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# loading NYC/TLC Yellow Cab dataset
df_trip_data = pd.read_parquet("data/trip-data.parquet")
df_trip_data.head()
```

```python
df_climate_data = pd.read_excel("data/weather-nyc.xlsx")
df_climate_data.head()
```

```python
# add column date the trip data
new_col = pd.DataFrame({"date": []}).astype('string')
df_trip_data = pd.concat([df_trip_data, new_col], axis=1)
```

```python

# extract date from pickup and add to new column "date"
for i, row in df_trip_data.iterrows():
    extracted_date = row.tpep_pickup_datetime.strftime("%Y-%m-%d")
    df_trip_data.at[i, "date"] = extracted_date 
```

## Merging the datasets

```python
# merge the the two datasets dropping the rows with NaN
merged_df = pd.merge(df_trip_data, df_climate_data, on="date", how="inner")
print("Index before merge",df_trip_data.index, "\n index after merge", merged_df.index )
```

```python
merged_df.isnull().sum()
```

```python
# drop columns with significant null values i.e. (over 6 milion) from the climate dataset
merged_df = merged_df.drop(columns=["prcp", "snow", "wdir", "wpgt", "tsun"], axis=1)
```

```python
merged_df.isnull().sum()
```

```python
# dealing with missing data
modify_df = merged_df                    # copy merged_df and modify
# for continous data fill the missing data with respective column mean
missing_data = ['passenger_count', 'congestion_surcharge']
modify_df[missing_data] = modify_df[missing_data].fillna(modify_df[missing_data].mean())
```

```python
# because weather conditions does not change drastically compared with previous observation forward fill the missing values
missing_data=['tavg', 'tmin', 'tmax', 'wspd', 'pres']
modify_df[missing_data] = modify_df[missing_data].ffill()
```

```python
modify_df.isnull().sum()
```

```python
# For rateCode it is quantitative data and RatecodeID == 1 is significant relative to other IDs therefore fillna with mode()
unique = {}
for rate in merged_df['RatecodeID'].unique(): 
    count = merged_df['RatecodeID'].eq(rate).sum()
    unique[rate] = count
print(unique)
modify_df["RatecodeID"] = merged_df["RatecodeID"].fillna(merged_df["RatecodeID"].mode().iloc[0])
```

```python
#For store_and_fwd_flag it is quantitative data and store_and_fwd_flag == N is significant relative to other IDs therefore fillna with mode()
unique = {}
for flag in merged_df['store_and_fwd_flag'].unique(): 
    count = merged_df['store_and_fwd_flag'].eq(flag).sum()
    unique[flag] = count
print(unique)
modify_df["store_and_fwd_flag"] = merged_df["store_and_fwd_flag"].fillna(merged_df["store_and_fwd_flag"].mode().iloc[0])
```

```python
# all entries in airport_fee is None therefore fillna with 0
unique = {}
for rate in merged_df['airport_fee'].unique():
    print(rate)
    count = merged_df['airport_fee'].eq(rate).sum()
    unique[rate] = count
print(unique)
# merged_df.columns
modify_df["airport_fee"] = merged_df["airport_fee"].fillna(value=0)
```

```python
modify_df.isnull().sum()
```

After handling all null values

```python
# rewrite merged to modifed dataframe
merged_df = modify_df
```

```python
for_describe = merged_df
for_describe = for_describe.drop(columns=["tpep_pickup_datetime", "tpep_dropoff_datetime", "date"], axis=1) # drop date columns
for_describe.describe()
```

### business objective and framing the problem
From the data frames. It prompts that the training supervision is a supervised learning task. Since each instance comes with the expected output i.e., the expected trip duration.

Since the model is meant to predict duration of trips it is therefore a regression task, specifically multiple regression problem due to the multiple features. It is also a univariate regression problem as we are interested in predicting a single duration per trip.

### Performance measure
As recommended loss functions; Mean Square Error (MSE) and Mean Absolute Error (MAE)
